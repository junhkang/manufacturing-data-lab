```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import json
import matplotlib as mpl
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

plt.rcParams['font.family'] = 'AppleGothic'  # Mac ÌïúÍ∏ÄÌè∞Ìä∏
mpl.rcParams['axes.unicode_minus'] = False

# Í≤ΩÎ°ú ÏÑ§Ï†ï
data_dir = Path("..") / "raw_data"
label_path = Path("..") / "labels" / "left_label.json"

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú
df_data = pd.read_csv(data_dir / "left_data.csv", header=None)
with open(label_path) as f:
    label_list = json.load(f)
df_label = pd.DataFrame(label_list, columns=["thickness"])
df_label["id"] = df_label.index
```

üîπ 1Îã®Í≥Ñ: Í∏∞Ï§ÄÍ∞í Î≤óÏñ¥ÎÇú Ï†ïÎèÑ Í≥ÑÏÇ∞
ÎëêÍªò Í∏∞Ï§Ä(1.5mm, 0.8mm)ÏóêÏÑú ÏñºÎßàÎÇò Î≤óÏñ¥ÎÇ¨ÎäîÏßÄÎ•º ÏàòÏπòÌôî
‚Üí Ïù¥ÌõÑÏóê Ï†ïÎ∞ÄÎèÑ/Ïò§Ï∞®Ïú®/Îì±Í∏âÌôî Îì±Ïóê ÏÇ¨Ïö© Í∞ÄÎä•

```python
def deviation(thickness, lower=0.8, upper=1.5):
    if lower <= thickness <= upper:
        return 0.0
    elif thickness < lower:
        return lower - thickness
    else:
        return thickness - upper

df_label["deviation"] = df_label["thickness"].apply(deviation)
df_label[["id", "thickness", "deviation"]].head()
```

üîπ  2Îã®Í≥Ñ: Îì±Í∏âÌôî (Grade A/B/C)
deviation ÏàòÏπò Í∏∞Î∞òÏúºÎ°ú OK/NG Ïô∏Ïóê A/B/C Îì±Í∏â Î∂ÄÏó¨

A: Í∏∞Ï§Ä ÏôÑÏ†Ñ Ï∂©Ï°± (Ïò§Ï∞® 0)

B: Í∏∞Ï§Ä Í∑ºÏ†ë (Ïò§Ï∞® < 0.05mm)

C: Í∏∞Ï§Ä Î≤óÏñ¥ÎÇ®

```python
def grade(deviation):
    if deviation == 0:
        return "A"
    elif deviation <= 0.05:
        return "B"
    else:
        return "C"

df_label["grade"] = df_label["deviation"].apply(grade)
df_label[["id", "thickness", "deviation", "grade"]].head()
```

üîπ 3Îã®Í≥Ñ: Í≥†Í∏â ÌíàÏßà ÎùºÎ≤® ÏÉùÏÑ±
grade + ÎëêÍªò ÏûêÏ≤¥Î•º Ï°∞Ìï©Ìï¥ÏÑú ÏµúÏ¢Ö ÌíàÏßà ÎùºÎ≤® ÏÉùÏÑ±
‚Üí Ïòà: A_OK, B_NG, C_NG Îì±

```python
def final_label(row):
    ok = "OK" if 0.8 < row["thickness"] < 1.5 else "NG"
    return f"{row['grade']}_{ok}"

df_label["label"] = df_label.apply(final_label, axis=1)
df_label[["id", "grade", "label"]].head()
```

üîπ 4Îã®Í≥Ñ : ÏãúÍ∞Ñ Ï¢ÖÏÜçÏÑ± Î∂ÄÏó¨
Ïòà: ÏÉòÌîå ID ÏàúÏúºÎ°ú Ïù¥Ï†Ñ Ï†úÌíà ÎåÄÎπÑ ÎëêÍªò Î≥ÄÌôîÎüâ Í≥ÑÏÇ∞ ‚Üí Ïù¥ÏÉÅÌÉêÏßÄ Í∞ÄÎä•

```python
df_label["delta_thickness"] = df_label["thickness"].diff().fillna(0).abs()
df_label[["id", "thickness", "delta_thickness"]].head()
```
